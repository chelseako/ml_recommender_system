# Functions  

This section has the similarity functions (eucildean, pearson, and cosine), which will be used to determine the similarity between vectors; the item rating estimator functions, which will be used to estimate user's ratings on unrated movies; and the evaluation functions, which will be used to compare the mean absolute error over all users on set aside test data between the standard estimator and SVD estimator functions. The svdEst function was created using the 206 SVD components (for the full dataset), and the svdEst2 function was created using the 40 SVD components (for the truncated dataset).

### Similarity functions

```python
def ecludSim(inA,inB):
    return 1.0 / (1.0 + la.norm(inA - inB))

def pearsSim(inA,inB):
    if len(inA) < 3 : return 1.0
    return 0.5 + 0.5 * corrcoef(inA, inB, rowvar = 0)[0][1]

def cosSim(inA,inB):
    num = float(inA.T * inB)
    denom = la.norm(inA)*la.norm(inB)
    return 0.5 + 0.5 * (num / denom)
```

### Item rating estimator functions

```python
# Function to get selected user's estimated rating on item using original dataset.

def standEst(dataMat, user, simMeas, item):
    n = shape(dataMat)[1]
    simTotal = 0.0; ratSimTotal = 0.0

    for j in range(n):  # n = number of movies
        userRating = dataMat[user,j]  # Get selected user's rating on j movie
        if userRating == 0: continue  # Skip items selected user did not rate

        # Produces array of indexes of all users that rated both the selected 
        # movie item and movie j.
        overLap = nonzero(logical_and(dataMat[:,item]>0, \
                                      dataMat[:,j]>0))[0]
        # If no users rated both items, similarity = 0
        if len(overLap) == 0: similarity = 0

        # Otherwise, gets similarity btw vector of all users' ratings on movie 
        # item and vector of users' ratings on movie j to determine how similar
        # movie item is to movie j.
        else: similarity = simMeas(dataMat[overLap,item], \
                                   dataMat[overLap,j])
        #print 'the %d and %d similarity is: %f' % (item, j, similarity)

        # Add the similarity value btw movie j and movie item.
        simTotal += similarity

        # Multiply similarity value by selected user's rating on movie j
        ratSimTotal += similarity * userRating
    if simTotal == 0: return 0

    # Divide rating similarity total by similarity total to get predicted 
    # user rating on movie item.
    else: return ratSimTotal/simTotal  
```

```python
# Function to get selected user's estimated rating on item using SVD.

def svdEst(dataMat, user, simMeas, item):

    n = shape(dataMat)[1]
    simTotal = 0.0; ratSimTotal = 0.0

    # Do SVD
    data=mat(dataMat)
    U,Sigma,VT = la.svd(data)
    Sig205 = mat(eye(205)*Sigma[:205]) # Arrange Sig205 into a diagonal matrix
    # Get transformed dataset (produces 1682 x 205 matrix)
    xformedItems = data.T * U[:,:205] * Sig205.I     
    
    for j in range(n):  # n = 1682

        # Get selected user's rating for movie j from original dataset.
        userRating = data[user,j]  
        
        # Skip items selected user did not rate
        if userRating == 0 or j==item: continue  

        # For rated items, get similarity between selected movie item and 
        # movie j using SVD xformedItems matrix.
        similarity = simMeas(xformedItems[item,:].T,\
                             xformedItems[j,:].T)  
        #print 'the %d and %d similarity is: %f' % (item, j, similarity)

        # Add up similarity between selected item and all other rated items
        simTotal += similarity

        # Multiply the similarity by the userRating and add to rating similarity total
        ratSimTotal += similarity * userRating
    if simTotal == 0: return 0

    # Return estimated rating on selected item
    else: return ratSimTotal/simTotal
```

```python
# Function to get selected user's estimated rating on item using SVD
# with users by movies matrix using 100 most rated movies.

def svdEst2(dataMat, user, simMeas, item):

    n = shape(dataMat)[1]
    simTotal = 0.0; ratSimTotal = 0.0

    # Do SVD
    data=mat(dataMat)
    U,Sigma,VT = la.svd(data)
    Sig40 = mat(eye(40)*Sigma[:40]) # Arrange Sig40 into a diagonal matrix
    # Get transformed dataset (produces 100 x 40 matrix)
    xformedItems = data.T * U[:,:40] * Sig40.I     
    
    for j in range(n):  # n = 1682

        # Get selected user's rating for movie j from original dataset.
        userRating = data[user,j]  
        
        # Skip items selected user did not rate
        if userRating == 0 or j==item: continue  

        # For rated items, get similarity between selected movie item and 
        # movie j using SVD xformedItems matrix.
        similarity = simMeas(xformedItems[item,:].T,\
                             xformedItems[j,:].T)  
        #print 'the %d and %d similarity is: %f' % (item, j, similarity)

        # Add up similarity between selected item and all other rated items
        simTotal += similarity

        # Multiply the similarity by the userRating and add to rating similarity total
        ratSimTotal += similarity * userRating
    if simTotal == 0: return 0

    # Return estimated rating on selected item
    else: return ratSimTotal/simTotal
```

### Evaluation functions

```python
def cross_validate_user(dataMat, user, test_ratio, estMethod=standEst, simMeas=pearsSim):
    # Function gets all items rated by specified user, 
    # gets test_size number of rated items, given test_ratio, and randomly selects
    # this many rated items to withhold. Iterates through withheld items and
    # gets predicted score using given estMethod.
  
    number_of_items = np.shape(dataMat)[1]  # Number of movies

    rated_items_by_user = np.array([i for i in range(number_of_items) if dataMat[user,i]>0])

    #print('RATED ITEMS BY USER')
    #print(rated_items_by_user)
  
    # Get number of movies in test set
    test_size = round(test_ratio * len(rated_items_by_user))

    # Select test_size number of selected user rated movies randomly
    test_indices = np.random.randint(0, len(rated_items_by_user), test_size)
 
    # Get indxes of selected user's rated movies withheld for test set
    withheld_items = rated_items_by_user[test_indices]
    #print('\nWITHHELD ITEMS')
    #print(withheld_items)

    # Make copy of original dataset of users' ratings
    original_user_profile = np.copy(dataMat[user])
    #print('\nORIGINAL USER PROFILE')
    #print(original_user_profile)

    # Change data matrix values for withheld items to 0
    dataMat[user, withheld_items] = 0 # So that the withheld test items is not used in the rating estimation below
  
    error_u = 0.0
    count_u = len(withheld_items)

    # Compute absolute error for user u over all test items
    for item in withheld_items:
        # Estimate rating on the withheld item
        estimatedScore = estMethod(dataMat, user, simMeas, item)
        error_u = error_u + abs(estimatedScore - original_user_profile[:,item])	

    # Now restore ratings of the withheld items to the user profile
    for item in withheld_items:
        dataMat[user, item] = original_user_profile[:,item]

    # Return sum of absolute errors and the count of test cases for this user
    # Note that these will have to be accumulated for each user to compute MAE
    return error_u, count_u
```

The test function takes in a data matrix, test ratio, estimator method(standest or SVD) and a similarity measure. The test function loops through each user and calls the cross_validate_user function, which calcualtes a test set based on the test_ratio of 0.2 or 20%. The function returns the mean absolute error for the entire data set.

```python
def test(dataMat, test_ratio, estMethod=standEst, simMeasure=cosSim):
    # Function iterates over all users and performs evaluation by calling the
    # cross_validate_user function on each user. Returns mean absolute error, 
    # which is the ratio of total error across all test cases to the total number 
    # of test cases, for all users

    M = dataMat.shape[0]  # Number of users
    N = dataMat.shape[1]  # Number of items/movies
    Ratings = mat(dataMat)  # Convert dataMat to np matrix.

    totError = 0
    totCount = 0
    
    for u in range(M):  # Iterate over all users

        if u % 100 == 0:
            print(u)  # Print function progress

        err, ct = cross_validate_user(dataMat=Ratings, user=u, 
                                      test_ratio=test_ratio, estMethod=estMethod,
                                      simMeas = simMeasure)

        totError += err
        totCount += ct

    MAE = totError/totCount

    print("Overall Mean Absolute Error = %0.3f" % (MAE))
```

# Recommender Application

The recommender application below prompts a user to rate at least 20 movies. The functions take that input, create a new vector, append that vector to the existing 100 movies data matrix, and produce estimated ratings for unrated items. The predicted ratings are sorted, and users are presented with the top N number of movie recommendations as requested.

### get_movie_info()

```python
def get_movie_info(id):
    '''Input: movie id reset as int
    Output: title, avg. rating, number of ratings'''
  
    title = movies100.loc[movies100['movie_id2']==id, 'movie_title'].iloc[0]
  
    rating = movies100.loc[movies100['movie_id2']==id, 'avgRating'].iloc[0]
  
    numRatings = movies100.loc[movies100['movie_id2']==id, 'numMovieRatings'].iloc[0]
  
    return title, rating, numRatings
```

```python
def getNewRatings(usersMoviesMatrix):
    'Gets list of new ratings for movie ID given matrix of movie titles with index ID'

    Ratings = []
    possible_ratings = ['0','1','2','3','4','5']

    print('On a scale of:\n')
    print('Bad\t\tNeutral\t\tGood')
    print('1\t2\t3\t4\t5')
    print('\nPlease rate the following movies:')
    print('\nPress 0 (zero) if you have not seen the movie.\n') 
  
    nonZeroLen=0

    for i in idxLst:  # Iterate through list of movie indexes as ints
        print('\nMovie rating {} out of 20.\n'.format(nonZeroLen+1))
        title, rating, numRatings = get_movie_info(i)
        print(title)

        rating = input()
    
        while rating not in possible_ratings:
            print('Rating must be from 1 to 5 or 0 if you have not seen the movie. Try again.')
            rating = input()

        Ratings.append((int(rating)))

        nonZeroLen = np.count_nonzero(Ratings)
    
        if nonZeroLen == 20:  

            vecLength = usersMoviesMatrix.shape[1]
            allRatings = getInputVector(Ratings, vecLength)
            return allRatings
```

### getInputVector()

```python
def getInputVector(ratingsLst, vecLength):
    'Appends 0s to unrated movies and returns full user rating vector'

    lstLen = len(ratingsLst)

    newLen = vecLength - lstLen

    for i in range(newLen):
        ratingsLst.append(0)

    allRatings = mat(ratingsLst)

    return allRatings
```

### recommend()

```python
def recommend(dataMat, userVec, N=5, simMeas=cosSim, estMethod=standEst):
    
    dataMat=np.array(dataMat)
    userVec = np.array(userVec)
    data = np.append(dataMat,userVec,axis=0)
  
    data=mat(data)

    userVec=mat(userVec)

    # List of indices of top 100 scores dataframe
    unratedItems = nonzero(userVec.A==0)[1]

    if len(unratedItems) == 0: 
        return 'You rated everything. No new recommendations.'
  
    itemScores = []
    for item in unratedItems:
        estimatedScore = estMethod(data, 943, cosSim, item)
        itemScores.append((item, estimatedScore))
  
    itemScores.sort(key=lambda x: x[1], reverse=True)
    top_movies = itemScores[:N]
  
    for i in range(len(top_movies)):
        predRating = round(top_movies[i][1], 2)
        print('\nRecommended movie', i+1, '\nPredicted rating:', predRating)
        title, rating, numRatings = get_movie_info(top_movies[i][0])
        print('Title: {}\tAverage Rating: {}\t Number of Ratings: {}'.format(title, rating, numRatings))

    return 
```

### main()

```python
def main(dataMat, N=5, simMeas = cosSim, estMethod=standEst):

    newUserVec = getNewRatings(dataMat)

    recommend(dataMat, newUserVec, N=N, simMeas=simMeas, estMethod=estMethod)
    return
```
